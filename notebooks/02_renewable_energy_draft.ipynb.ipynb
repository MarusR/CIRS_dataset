# Fixed-Effects Regression Analysis

This notebook performs fixed-effects regression analysis on the World Development Indicators (WDI) dataset to explore the relationship between renewable energy use, air quality, and life expectancy.

```python
# Import necessary libraries
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style='whitegrid')
```

## Load Data

Load the cleaned dataset from the processed data directory.

```python
# Load the processed dataset
data = pd.read_csv('../data/processed/cleaned_wdi_data.csv')

# Display the first few rows of the dataset
data.head()
```

## Data Preparation

Prepare the data for fixed-effects regression analysis by ensuring the necessary variables are present and handling any missing values.

```python
# Check for missing values
missing_values = data.isnull().sum()
missing_values[missing_values > 0]
```

## Fixed-Effects Model

Define and fit the fixed-effects regression model.

```python
# Define the fixed-effects model
model = ols('life_expectancy ~ renewables_share + pm25 + gdp_per_capita + C(country)', data=data).fit()

# Display the model summary
model_summary = model.summary()
model_summary
```

## Model Diagnostics

Evaluate the model diagnostics to check for assumptions and potential issues.

```python
# Plot residuals
plt.figure(figsize=(10, 6))
sns.residplot(x=model.fittedvalues, y=model.resid, lowess=True, 
               line_kws={'color': 'red', 'lw': 1})
plt.xlabel('Fitted values')
plt.ylabel('Residuals')
plt.title('Residuals vs Fitted Values')
plt.axhline(0, linestyle='--', color='red')
plt.show()
```
