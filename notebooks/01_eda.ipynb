# Exploratory Data Analysis (EDA)

This notebook is dedicated to performing exploratory data analysis on the World Development Indicators (WDI) dataset. The goal is to understand the data, identify patterns, and prepare for further analysis.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style='whitegrid')
```

```python
# Load the dataset
data_path = '../data/processed/your_cleaned_data.csv'  # Update with the actual path
wdi_data = pd.read_csv(data_path)

# Display the first few rows of the dataset
wdi_data.head()
```

```python
# Summary statistics
wdi_data.describe()
```

```python
# Check for missing values
missing_values = wdi_data.isnull().sum()
missing_values[missing_values > 0]
```

```python
# Visualize missing data
plt.figure(figsize=(12, 6))
sns.heatmap(wdi_data.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(10, 8))
correlation_matrix = wdi_data.corr()
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Visualize key relationships
plt.figure(figsize=(10, 6))
sns.scatterplot(data=wdi_data, x='GDP per capita', y='Life expectancy', hue='Region')
plt.title('Life Expectancy vs GDP per Capita')
plt.xlabel('GDP per Capita')
plt.ylabel('Life Expectancy')
plt.legend(title='Region')
plt.show()
```

```python
# Save the cleaned data for further analysis
wdi_data.to_csv('../data/processed/cleaned_wdi_data.csv', index=False)
```
